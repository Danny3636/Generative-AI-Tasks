**This code demonstrates how to use PyTorch for automatic differentiation, specifically how to define a simple mathematical function with requires_grad=True tensors and then compute gradients for these tensors using the .backward() method.**

```python
import torch

# Given Initial Values
w_val = 2.0
x_val = 4.0
b_val = 1.5

# Create tensors for w, x, and b with requires_grad=True
w = torch.tensor(w_val, requires_grad=True)
x = torch.tensor(x_val, requires_grad=True)
b = torch.tensor(b_val, requires_grad=True)

# Define the function: z = sigmoid(w * x^2) + 1 / b^3
# where sigma is the sigmoid function, which in PyTorch is torch.sigmoid()
intermediate = w * (x**2)
sigmoid_output = torch.sigmoid(intermediate)
power_b = b**3
reciprocal_b = 1 / power_b
z = sigmoid_output + reciprocal_b

# Call the .backward() method to compute the gradients
z.backward()

# Print the computed gradients for w, x, and b
print(f"Gradient of z with respect to w: {w.grad}")
print(f"Gradient of z with respect to x: {x.grad}")
print(f"Gradient of z with respect to b: {b.grad}")
```
